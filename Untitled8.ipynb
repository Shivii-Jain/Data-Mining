```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.ensemble import IsolationForest
from scipy.cluster.hierarchy import linkage, fcluster

df = pd.read_csv(
    "Warehouse_and_Retail_Sales.csv",
    engine="python",
    on_bad_lines="skip",
    usecols=lambda c: any(
        k in c.lower()
        for k in ["year", "month", "item", "retail", "warehouse"]
    )
)

df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

df = df.rename(columns={"warehouse_sales": "wholesale_sales"})

if "item_type" in df.columns:
    df["item_type"] = df["item_type"].astype("category")

for col in ["retail_sales", "retail_transfers", "wholesale_sales"]:
    if col in df.columns:
        df[col] = df[col].astype("float32")

df = df.dropna(subset=["retail_sales", "retail_transfers", "wholesale_sales"])

df = df[
    (df["retail_sales"] >= 0) &
    (df["retail_transfers"] >= 0) &
    (df["wholesale_sales"] >= 0)
]

print("After cleaning:", df.shape)

df["total_sales"] = (
    df["retail_sales"] +
    df["retail_transfers"] +
    df["wholesale_sales"]
)

df = df[df["total_sales"] > 0]

df["wholesale_ratio"] = df["wholesale_sales"] / df["total_sales"]
df["retail_ratio"] = df["retail_sales"] / df["total_sales"]
df["log_total_sales"] = np.log1p(df["total_sales"])

X = df[["wholesale_ratio", "retail_ratio", "log_total_sales"]]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

del X

wcss = []
for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    wcss.append(km.inertia_)

# Elbow Method Plot
plt.figure()
plt.plot(range(2, 7), wcss, marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method for K-Means")
plt.show()

# Final K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["kmeans_cluster"] = kmeans.fit_predict(X_scaled)

print("\nK-Means cluster sizes:")
print(df["kmeans_cluster"].value_counts())


# Cluster size distribution
plt.figure()
df["kmeans_cluster"].value_counts().sort_index().plot(kind="bar")
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.title("K-Means Cluster Size Distribution")
plt.show()

# Wholesale ratio by cluster
plt.figure()
df.boxplot(column="wholesale_ratio", by="kmeans_cluster")
plt.xlabel("Cluster")
plt.ylabel("Wholesale Ratio")
plt.title("Wholesale Ratio by K-Means Cluster")
plt.suptitle("")
plt.show()

# Scatter plot
plt.figure()
for c in sorted(df["kmeans_cluster"].unique()):
    sub = df[df["kmeans_cluster"] == c]
    plt.scatter(
        sub["wholesale_ratio"],
        sub["log_total_sales"],
        alpha=0.4,
        label=f"Cluster {c}"
    )

plt.xlabel("Wholesale Ratio")
plt.ylabel("Log(Total Sales)")
plt.title("K-Means Clustering Visualization")
plt.legend()
plt.show()



# HIERARCHICAL CLUSTERING
sample_df = df.sample(n=20000, random_state=42)

X_sample = sample_df[["wholesale_ratio", "retail_ratio", "log_total_sales"]]
X_sample_scaled = StandardScaler().fit_transform(X_sample)

Z = linkage(X_sample_scaled, method="ward")
sample_df["hier_cluster"] = fcluster(Z, 3, criterion="maxclust")

print("\nHierarchical clustering (sample) sizes:")
print(sample_df["hier_cluster"].value_counts())

plt.figure()
for c in sorted(sample_df["hier_cluster"].unique()):
    sub = sample_df[sample_df["hier_cluster"] == c]
    plt.scatter(
        sub["wholesale_ratio"],
        sub["log_total_sales"],
        alpha=0.5,
        label=f"Cluster {c}"
    )

plt.xlabel("Wholesale Ratio")
plt.ylabel("Log(Total Sales)")
plt.title("Hierarchical Clustering (Sample)")
plt.legend()
plt.show()



# GAUSSIAN MIXTURE MODEL
gmm = GaussianMixture(n_components=3, random_state=42)
df["gmm_cluster"] = gmm.fit_predict(X_scaled)

print("\nGMM cluster sizes:")
print(df["gmm_cluster"].value_counts())

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
df["kmeans_cluster"].value_counts().sort_index().plot(kind="bar")
plt.title("K-Means Cluster Sizes")
plt.xlabel("Cluster")
plt.ylabel("Count")

plt.subplot(1,2,2)
df["gmm_cluster"].value_counts().sort_index().plot(kind="bar")
plt.title("GMM Cluster Sizes")
plt.xlabel("Cluster")
plt.ylabel("Count")

plt.tight_layout()
plt.show()

# ANOMALY DETECTION (ISOLATION FOREST)

iso = IsolationForest(
    contamination=0.02,
    random_state=42,
    n_estimators=100
)

df["anomaly"] = iso.fit_predict(X_scaled)

anomalies = df[df["anomaly"] == -1]
normal = df[df["anomaly"] == 1]

print("\nAnomalies detected:", anomalies.shape[0])
print("Anomaly rate:", anomalies.shape[0] / df.shape[0])

plt.figure()
plt.scatter(
    normal["wholesale_ratio"],
    normal["log_total_sales"],
    alpha=0.3,
    label="Normal"
)

plt.scatter(
    anomalies["wholesale_ratio"],
    anomalies["log_total_sales"],
    alpha=0.9,
    label="Anomaly"
)

plt.xlabel("Wholesale Ratio")
plt.ylabel("Log(Total Sales)")
plt.title("Isolation Forest Anomaly Detection")
plt.legend()
plt.show()

print("\nK-Means Cluster Profile:")
print(
    df.groupby("kmeans_cluster")[
        ["retail_sales", "wholesale_sales", "total_sales"]
    ].mean()
)

print("\nGMM Cluster Profile:")
print(
    df.groupby("gmm_cluster")[
        ["retail_sales", "wholesale_sales", "total_sales"]
    ].mean()
)
```
